```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split

```

# Autoencoder

```python
from keras.layers import Input, Dense, Flatten, Reshape
from keras.models import Model, load_model, save_model
```

```python
Specs.shape
```

```python
def create_deep_dense_ae():
    # Размерность кодированного представления
    encoding_dim = 6
    activation = 'selu'

    # Энкодер
    input_spec = Input(shape=(36, 1))
    flat_spec = Flatten()(input_spec)
    x = Dense(encoding_dim*4, activation=activation)(flat_spec)
    x = Dense(encoding_dim*3, activation=activation)(x)
    x = Dense(encoding_dim*2, activation=activation)(x)
    encoded = Dense(encoding_dim, activation='linear')(x)
    
    # Декодер
    input_encoded = Input(shape=(encoding_dim,))
    x = Dense(encoding_dim*2, activation=activation)(input_encoded)
    x = Dense(encoding_dim*3, activation=activation)(x)
    x = Dense(encoding_dim*4, activation=activation)(x)
    flat_decoded = Dense(36, activation='sigmoid')(x)
    decoded = Reshape((36, 1))(flat_decoded)
    
    # Модели
    encoder = Model(input_spec, encoded, name="encoder")
    decoder = Model(input_encoded, decoded, name="decoder")
    autoencoder = Model(input_spec, decoder(encoder(input_spec)), name="autoencoder")
    return encoder, decoder, autoencoder

d_encoder, d_decoder, d_autoencoder = create_deep_dense_ae()
d_autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
encoder, decoder, autoencoder = create_deep_dense_ae()
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
autoencoder.summary()
```

```python
Spec_Data = Data.filter(regex='^Ch')
```

```python
arr = []
for spec in Spec_Data.values:
    arr.append(spec.reshape(36, 1))
    
arr = np.array(arr)

X_train, X_test, _,_ = train_test_split(arr, arr, test_size=0.2, random_state = 42)
```

```python
autoencoder.fit(X_train, X_train,
                epochs=500,
                batch_size=100,
                shuffle=True,
                verbose=1,
                validation_data=(X_test, X_test))
```

```python
n = 10

specs = X_test[:n]
encoded_specs = encoder.predict(specs, batch_size=n)
decoded_specs = decoder.predict(encoded_specs, batch_size=n)

fig, axes = plt.subplots(10,1)
fig.set_figheight(80)
fig.set_figwidth(12)

for i in range(n):
    axes[i].plot(specs[i], label='Initial Spectra')
    axes[i].plot(decoded_specs[i], label='Decoded Spectra')
    axes[i].grid()
    axes[i].legend()
```
