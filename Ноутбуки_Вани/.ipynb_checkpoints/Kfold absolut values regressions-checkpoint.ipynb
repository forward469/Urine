```python
import pandas as pd
import numpy as np
import sklearn

from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
analysis = pd.read_excel('20.06.25 список образцов Моча.xlsx')

new_header = analysis.iloc[0]
analysis = analysis[1:]
analysis.columns = new_header
analysis.drop(analysis.index[0], inplace=True)

analysis
```

```python
spectra = pd.read_excel('pi_cup.xlsx')

spectra.drop(columns=['Unnamed: 0'], inplace=True)
# Удаляем все спектры, соответствующие Dataset = 'Chm38', потому что в исходной таблице нет результатов по этому образцу
spectra = spectra.drop(spectra.loc[spectra['Dataset'] == 'Chm38'].index).reset_index(drop=True)

spectra
```

```python
#Забираю только спектральные данные + Dataset
spectra = spectra.iloc[:,0:37]

spectra
```

```python
print('Всего analysis:','\t\t', analysis.drop_duplicates('Порядковый номер').index.nunique())
print('Всего spectra:','\t\t', spectra.drop_duplicates('Dataset').index.nunique())
```

```python
dataset_list = spectra.drop_duplicates('Dataset').Dataset.reset_index()
dataset_list.drop(columns='index', inplace = True)
dataset_list = dataset_list.Dataset
dataset_list
```

```python
type(dataset_list)
```

```python
#Забираем данные абсолютных значений
for i in range(len(dataset_list)):
    spectra.loc[spectra['Dataset'] == dataset_list[i], 'Leukocyte_cytometry_values'] = analysis.loc[analysis['Порядковый номер'] == dataset_list[i]]['Лейкоциты цитометрия'].value_counts().index[0]
    spectra.loc[spectra['Dataset'] == dataset_list[i], 'Bacteria_values'] = analysis.loc[analysis['Порядковый номер'] == dataset_list[i]]['Бактерии'].value_counts().index[0]
```

```python
spectra
```

```python
def pls_func(X_train, y_train, X_test, y_test, n_comp):
  pls = PLSRegression(n_components=n_comp)
  pls.fit(X_train,y_train)
  pls_result = pls.predict(X_test)
  y_pred = pd.DataFrame(pls_result)
  y_pred.columns = list(y_test)

  return y_pred
```

```python
def split_data(X, y, groups):
  train_groups, test_groups = train_test_split(groups, test_size = 0.2)
  X_train = X[X.Dataset.isin(train_groups)].copy()
  X_train.drop(columns = 'Dataset', inplace = True)
  X_test = X[X.Dataset.isin(test_groups)].copy()
  X_test.drop(columns = 'Dataset', inplace = True)
  y_train = y[y.Dataset.isin(train_groups)].copy()
  y_train.drop(columns = 'Dataset', inplace = True)
  y_test = y[y.Dataset.isin(test_groups)].copy()
  y_test.drop(columns = 'Dataset', inplace = True)

  return X_train, y_train, X_test, y_test
```

```python
def calculate_metrics(y_test, y_pred, n_comp, iteration = 0):
    metrics_list = []
    metrics_list.append(['n_comp','Parameter','MSE','MAPE','iteration'])
    mse = mean_squared_error(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    if isinstance(y_test, pd.DataFrame):
      metrics_list.append([n_comp,y_test.columns[0],mse,mape,iteration])
    else:
      metrics_list.append([n_comp,y_test.name,mse,mape,iteration])
    metrics_data = pd.DataFrame(metrics_list)
    new_header = metrics_data.iloc[0]
    metrics_data = metrics_data[1:]
    metrics_data.columns = new_header

    return metrics_data
```

```python
def by_parameter_mean(metrics_data):
    metrics_data_mean = pd.DataFrame()
    parameters_list = metrics_data.drop_duplicates('Parameter').Parameter
    parameters_list = parameters_list.reset_index(drop=True)
    for i in range(len(parameters_list)):
      for j in range(29):
        to_concat = pd.DataFrame(metrics_data.loc[metrics_data['Parameter']==parameters_list[i]].loc[metrics_data['n_comp']==j+2].mean()).T
        to_concat['Parameter'] = parameters_list[i]
        to_concat.drop(columns = 'iteration', inplace = True)
        metrics_data_mean = pd.concat([metrics_data_mean, to_concat])
        metrics_data_mean = metrics_data_mean.reset_index(drop=True)

    return metrics_data_mean
```

```python
X_data = spectra.iloc[:,0:37]
y_data = spectra.iloc[:,36:]
groups = spectra.Dataset.drop_duplicates()
```

```python
kf = KFold()
iteration = 1
metrics_data_PLS_two = pd.DataFrame()
for train_groups_index, test_groups_index in kf.split(groups):
  train_groups, test_groups = groups.iloc[train_groups_index], groups.iloc[test_groups_index]
  X_train = X_data[X_data.Dataset.isin(train_groups)].copy()
  X_train.drop(columns = 'Dataset', inplace = True)
  X_test = X_data[X_data.Dataset.isin(test_groups)].copy()
  X_test.drop(columns = 'Dataset', inplace = True)
  y_train = y_data[y_data.Dataset.isin(train_groups)].copy()
  y_train.drop(columns = 'Dataset', inplace = True)
  y_test = y_data[y_data.Dataset.isin(test_groups)].copy()
  y_test.drop(columns = 'Dataset', inplace = True)
  for i in range(29):
    n_comp = 2 + i
    y_pred = pls_func(X_train, y_train, X_test, y_test, n_comp)
    y_list = list(y_pred)
    for j in range(len(y_list)):
      to_concat = calculate_metrics(y_test.loc[:, y_list[j]], y_pred.loc[:, y_list[j]], n_comp, iteration)
      metrics_data_PLS_two = pd.concat([metrics_data_PLS_two,to_concat])
      metrics_data_PLS_two = metrics_data_PLS_two.reset_index(drop = True)
  iteration = iteration + 1
```

```python
metrics_data_PLS_two_mean = by_parameter_mean(metrics_data_PLS_two)
metrics_data_PLS_two_mean
```

```python
print('Минимум mean squared error для Leukocyte_cytometry')
print('(PLS Leukocyte_cytometry & Bacteria)')

display(metrics_data_PLS_two_mean.loc[metrics_data_PLS_two_mean['Parameter'] == 
        'Leukocyte_cytometry_values'].loc[metrics_data_PLS_two_mean['MSE']
         == metrics_data_PLS_two_mean.loc[metrics_data_PLS_two_mean['Parameter']
        =='Leukocyte_cytometry_values'].MSE.min()])

print('\n\nМинимум mean squared error для Bacteria')
print('(PLS Leukocyte_cytometry & Bacteria)')

display(metrics_data_PLS_two_mean.loc[metrics_data_PLS_two_mean['Parameter'] ==
        'Bacteria_values'].loc[metrics_data_PLS_two_mean['MSE'] ==
        metrics_data_PLS_two_mean.loc[metrics_data_PLS_two_mean['Parameter'] 
          =='Bacteria_values'].MSE.min()])
```
