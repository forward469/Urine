## Счетчики по аномалиям

```python
from tqdm.notebook import tqdm

Anoms = []
Norms = []

for anom in tqdm(Anom_names):
    Counts = Data[anom].value_counts()
    Norms.append(Counts[0])
    try:
        Anoms.append(Counts[1])
    except KeyError:
        Anoms.append(0)

Stats = pd.DataFrame(columns=["Anom_Name", "Anom_Count", "Norm_Count"])
Stats["Anom_Name"] = list(map(lambda x: EN_Anom_Names[x], Anom_names))
Stats["Anom_Count"] = Anoms
Stats["Norm_Count"] = Norms
Stats["Total"] = Stats.Anom_Count + Stats.Norm_Count
```

<div class = "alert alert-block alert-warning">
<b>Формулы подсчета</b>

* Количество больных пациентов по результатам анализов - $N_{\text{anom}}$
* Количество здоровых пациентов по результатам анализов - $N_{\text{norm}}$
* Количество выявленных моделью как больные - $M_{\text{anom}}$
* Количество выявленных моделью как здоровые - $M_{\text{norm}}$
    
* Метрика модели "Специфичность" : $\operatorname{SPEC} = \operatorname{Specificity} = \frac{TN}{TN + FP} =
\frac{M_{\text{norm}}}{N_{\text{norm}}}$
* Метрика модели "Чувствительность" : $\operatorname{SENS} = \operatorname{Sensitivity} = \frac{TP}{TP + FN} = 
\frac{M_{\text{anom}}}{N_{\text{anom}}}$


</div>

### По результатам модели

```python
MODEL_RESULT = pd.read_csv("../Results/RESULT.csv")
```

```python
def my_fbeta_score(sens, spec, beta=1):
    return (1 + beta**2) * (spec * sens) / (beta**2 * spec + sens)
```

```python
MODEL_RESULT['harm_avg'] = my_fbeta_score(MODEL_RESULT.sens.values, MODEL_RESULT.spec.values)
grouped = MODEL_RESULT.groupby('Anomaly').max().harm_avg.dropna()
```

```python
SENS, SPEC = [], []

for name, score in zip(grouped.index, grouped.values):
    SENS.append(MODEL_RESULT[MODEL_RESULT.harm_avg == score].sens.values[0])
    SPEC.append(MODEL_RESULT[MODEL_RESULT.harm_avg == score].spec.values[0])

sample = pd.DataFrame(columns=["Name", "Sens", "Spec"])
sample["Name"] = grouped.index
sample["Sens"] = SENS
sample["Spec"] = SPEC
```

```python
sample['EN_name'] = list(map(lambda x: EN_Anom_Names[x], sample.Name))
```

```python
Stats_with_results = pd.merge(left=Stats,
                              right=sample,
                              left_on="Anom_Name",
                              right_on="EN_name").drop(['Name', 'EN_name'],
                                                       axis=1)

Stats_with_results = Stats_with_results.iloc[:-1, :]  # Выкинул сперматозоиды из визуализации
Stats_with_results.sort_values(by="Anom_Count", ascending=False, inplace=True)
```

## Intro

Перед тем как приступить к обзору технических деталей решения задачи определения положительных результатов медицинского анализа у пациента, необходимо поставить задачу в рамках терминологии машинного обучения. Пусть задано множество объектов $X$, множество _допустимых_ ответов $Y$ , и сущeствует _целевая функция_ (target function) $y^* : X \rightarrow Y$, значения которой $y_i = y^∗(xi)$ известны только на конечном подмножестве объектов $\{x_1, \ldots , x_l\} \in X$. Пары «объект–ответ» $(x_i, y_i)$ называются _прецедентами_ (sample). Совокупность пар $X^l = (xi, yi)^l_{i=1}$ называется _обучающей выборкой_ (training sample).

Задача _обучения по прецедентам_ заключается в том, чтобы по выборке $X_l$ восстановить зависимость $y^∗$, то есть построить _решающую функцию_ (decision function) $a: X \rightarrow Y$ , которая приближала бы целевую функцию $y^∗(x)$, причём не только на объектах обучающей выборки, но и на всём множестве $X$.

Решающая функция должна допускать эффективную компьютерную реализацию. По этой причине будем называть её алгоритмом.

В рамках нашего устройства, спектр представляет из себя два массива чисел длины 18 измеренных 18-ти канальным спектрометром с помощью двух разных источников света.

Следовательно, в качестве входных данных используется матрица вещественных чисел $X \in \mathbb{R}^{N\times36}$ , где N - количество пациентов в выборке.

А в качестве целевых переменных выступают бинарные параметры клинического анализа, где $0$ - означает что пациент не имеет патологии по данному параметру, а $1$ - означает, что пациент получил положительный результат по параметру. Матрицу целевых переменных будем обозначать $Y \in \mathbb{R}^{N\times P}$ , где $P$ - количество исследуемых клинических параметров.

Таким образом, переходя к задаче в терминах машинного обучения сформулируем задачу мультиразмерной бинарной классификации следующим образом : 

<div class="alert alert-block alert-info">
    
<b>Постановка задачи классификации</b>
    
Пусть $X$ — множество описаний объектов, $Y$  — множество наименований классов. Существует неизвестная целевая зависимость -- отображение $y^∗ : X \rightarrow Y$, значения которой известны только на объектах конечной обучающей выборки $X^m = {(x_1, y_1), \ldots, (x_m, y_m)}$. Требуется построить алгоритм $a : X \rightarrow Y$, способный классифицировать произвольный объект $x \in X$.
    
</div>

## Выбор метрик

После формальной постановки задачи, требуется выбрать ряд метрик для того, чтобы оценивать качество построенной модели независимо от выбора типа модели, ее набора гиперпараметров и способа предобработки данных.

Обычно, в случае медицинских тестов используются такие метрики качества как _специфичность_ и _чувствительность_. Введем формальные определения эти понятиям :

<div class = "alert alert-block alert-info">
    
> <b>Чувствительность</b> - представляет собой вероятность положительного результата теста среди пациентов при наличии заболевания (истинно-положительная частота)
    
$$\operatorname{SENS} = \frac{\operatorname{TP}}{\operatorname{TP} + \operatorname{FN}}$$
    
</div>


<div class = "alert alert-block alert-info">
    
> <b>Специфичность</b> - представляется собой вероятность отрицательных результатов теста среди пациентов при отсутствии заболевания (истинно-отрицательная частота)
    
$$\operatorname{SPEC} = \frac{\operatorname{TN}}{\operatorname{TN} + \operatorname{FP}}$$
    
</div>

Большой преимущество выбранных метрик заключается в их явной вероятностной интерпретации, что полезно. Но для прямого сравнения двух разных моделей при выборе лучшей, два значения метрик всегда хуже чем одно, но при этом важны обе. Поэтому введем дополнительно метрику $\text{F-score}$ равную взвешенному гармоническому среднему _Чувствительности_ и _Специфичности_ : 

$$F_{\beta}\text{-score} = \left(1+\beta^{2}\right) \cdot \frac{\text { specificity } \cdot \text { sensitivity }}{\left(\beta^{2} \cdot\right. \text { specificity })+\text { sensitivity }}$$


## Pipeline

Основная идея нашей модели классификации заключается в предварительной категоризации обучающей выборки по параметрам медицинского анализа. ### PLS algorithm

Коренной моделью является модель _Partial Least Squares (PLS)_. _Частичная регрессия наименьших квадратов (PLS)_ - это статистический метод, который имеет некоторое отношение к _регрессии главных компонентов_; вместо поиска гиперплоскостей максимальной дисперсии между ответом и независимыми переменными он находит модель линейной регрессии, проецируя предсказанные переменные и наблюдаемые переменные в новое пространство. 

PLS используется для нахождения фундаментальных отношений между двумя матрицами ($X$ и $Y$), то есть скрытого переменного подхода к моделированию ковариационных структур в этих двух пространствах. Модель PLS попытается найти многомерное направление в пространстве $X$, которое объясняет максимальное направление многомерной дисперсии в пространстве $Y$. Регрессия PLS особенно подходит, когда существует мультиколлинеарность между значениями $X$. В случае нашей задачи, многомерной матрицей ответов $Y$ является бинарная матрица, соответствующая положительным и отрицательным (1 и 0) результатам медицинских анализов, где каждая строка матрицы соответствует пациенту, а каждый столбец соответствует наименованию параметра.

По сути, это разбиение можно трактовать как разбиение по характерным размерам, но в тоже время есть и функциональный смысл каждой категории.

### Classifier

Таким образом мы получаем 18 отдельных классификаторов, каждый из которых обучен предсказывать бинарную метку по поступающему на вход спектру пропускания образца. В параметрах каждого такого классификатора помимо информации о предсказываемой метке содержится информация так же и о всех метках параметров соответствующей категории нашего разбиения.

## Intro

Before starting to review the technical details of solving the problem of determining positive results of a medical analysis in a patient, it is necessary to pose the problem in terms of machine learning terminology. Namely, to determine the strict formulation of the problem of multiclass classification, to determine the error functions and metrics for assessing the quality of the model, the choice of the type of model, and its modifications.

But first, let's introduce a few terms to understand the problem statement in terms of machine learning terms.

Let a set of objects $ X $ be given, a set of _admissible_ answers $ Y $, and there is a _target function_ $ y^*: X \rightarrow Y $, the values of which $ y_i = y^∗(xi) $ are known only on the finite a subset of objects $ \{x_1, \ldots, x_l \} \in X $. The pairs "object-target" $ (x_i, y_i) $ are called _precedents_ (_samples_). The collection of pairs $ X^l = (xi, yi) ^l_{i = 1} $ is called the _training sample_.

The task of _learning by precedents_ is to reconstruct the $ y^∗ $ dependence from the $ X_l $ sample, that is, to construct a decision function $ a: X \ rightarrow Y $, which would approximate the objective function $ y ^ ∗ ( x) $, and not only on the objects of the training set, but also on the entire set $ X $.

The decisive function must allow for efficient computer implementation. For this reason, we will call it an algorithm.

## Постановка задачи

The essence of the problem is to determine the binary parameters of a clinical test in patients from the measured optical transmission spectra of urine samples, respectively. Within the framework of our device, the spectrum is two arrays of numbers of length 18 measured with an 18-channel spectrometer using two different light sources.

Therefore, the matrix of real numbers $ X \in \mathbb {R}^{N \times 36} $ is used as input, where N is the number of patients in the sample.

And the target variables are binary parameters of clinical analysis, where $0$ - means that the patient has no pathology for this parameter, and $1$ - means that the patient received a positive result for the parameter. The target variable matrix will be denoted by $Y \in \mathbb {R}^{N \times P} $, where $P$ is the number of clinical parameters under study.

Thus, passing to the problem in terms of machine learning, we will formulate the multidimensional binary classification problem as follows:

<div class = "alert alert-block alert-info">
    
<b> Statement of the classification problem </b>
    
Let $X$ be the set of object descriptions, $Y$ the set of class names. There is an unknown target dependence - the mapping $ y^∗ : X \rightarrow Y $, the values of which are known only on the objects of the finite training set $ X^m = {(x_1, y_1), \ldots, (x_m, y_m)} $. It is required to construct an algorithm $ a: X \rightarrow Y $, capable of classifying an arbitrary object $ x \in X $.
    
</div>

## Выбор метрик

After the formal statement of the problem, it is required to select a number of metrics in order to assess the quality of the constructed model, regardless of the choice of the model type, its set of hyperparameters and the method of data preprocessing.

Typically, in the case of medical tests, quality metrics such as _specificity_ and _sensitivity_ are used. Let us introduce formal definitions of these concepts:

<div class = "alert alert-block alert-info">
    
> <b>Sensitivity</b> - represents the probability of a positive test result among patients with a disease (true-positive rate)
    
$$ \operatorname {SENS} = \frac{\operatorname {TP}}{\operatorname{TP} + \operatorname{FN}} $$
    
</div>


<div class = "alert alert-block alert-info">
    
> <b>Specificity</b> - represents the probability of negative test results among patients in the absence of disease (true negative rate)
    
$$ \operatorname{SPEC} = \frac{\operatorname {TN}}{\operatorname{TN} + \operatorname{FP}} $$
    
</div>

A big advantage of the chosen metrics lies in their explicit probabilistic interpretation, which is useful. But for direct comparison of two different models when choosing the best one, two metric values are always worse than one, but both are important. Therefore, we introduce additionally the $ \text{F-score}$ metric equal to the weighted harmonic mean of _Sensitivity_ and _Specificity_:

$$ F_{\beta}\text{-score} = \left(1+\beta^{2} \right) \cdot \frac{\text{specificity}\cdot \text{sensitivity}} {\left ( \beta ^ {2} \cdot \right. \text{specificity}) + \text{sensitivity}} $$

## Pipeline

The main idea of our classification model is to pre-categorize the training sample according to the parameters of medical analysis. In practice, this categorization allows you to accumulate and save information from the analysis results for different parameters in order to improve the quality of model responses for each of the parameters. Exactly how this improves quality will be described below in the _PLS algorithm_ part.

### PLS algorithm

The root model is the _Partial Least Squares (PLS) _ model. _Partial Least Squares Regression (PLS) _ is a statistical technique that has something to do with _ principal component regression_; instead of looking for the hyperplanes of maximum variance between the answer and the independent variables, it finds a linear regression model by projecting the predicted variables and observed variables into a new space.

PLS is used to find the fundamental relationship between two matrices ($ X $ and $ Y $), i.e. a hidden variable approach to modeling covariance structures in these two spaces. The PLS model will try to find the multidimensional direction in the $ X $ space that explains the maximum direction of the multidimensional variance in the $ Y $ space. PLS regression is especially suitable when there is multicollinearity between the values of $ X $. In the case of our problem, the multidimensional matrix of responses $ Y $ is a binary matrix corresponding to positive and negative (1 and 0) results of medical analyzes, where each row of the matrix corresponds to a patient, and each column corresponds to the name of the parameter.

### Categorization

The idea of categorization by the parameters of the $ Y $ response matrix is to try to transfer and save information from a group of parameters to improve the quality of model responses in the case of each specific parameter. We have broken down the parameters as follows:

- __Integral characteristics__: Density and pH.

- __Organic substances__: Proteins, glucose, ketones, etc.

- __Biological substances__: Erythrocytes, leukocytes, flat cells, etc.

In fact, this division can be interpreted as a division according to characteristic sizes, but at the same time there is a functional meaning for each category.

### Classifier

Thus, we get 18 separate classifiers, each of which is trained to predict a binary label from the input transmission spectrum of the sample. In the parameters of each such classifier, in addition to information about the predicted label, there is also information about all the labels of the parameters of the corresponding category of our partition.

For each parameter, honey. analysis, the selection of model hyperparameters was carried out using the cross-validation method maximizing the metric $ F_{\beta}\text{-score} $ - the harmonic average between Sensitivity and Specificity.

After that, all classifiers are combined into a common model in order to immediately receive a single response for the patient using the incoming spectrum: an array of binary responses for each of the parameters.

[Источник: ПОНИМАНИЕ МЕДИЦИНСКИХ ТЕСТОВ И ИХ РЕЗУЛЬТАТОВ](https://www.msdmanuals.com/ru/профессиональный/специальные-темы/особенности-принятия-клинического-решения/понимание-медицинских-тестов-и-их-результатов)

## Метрики

Чувствительность, специфичность и прогностическая ценность, как правило, считаются характеристиками самого теста, независимыми от популяции пациентов.

<div class = "alert alert-block alert-info">
    
* <b>Чувствительность</b>  – представляет собой вероятность положительного результата теста среди пациентов при наличии заболевания (истинно-положительная частота)
</div>

Таким образом, тест, который положителен у 8 из 10 пациентов при наличии заболевания имеет чувствительность 0,8 (также значение выражается как 80%). Чувствительность представляет, насколько хорошо тест обнаруживает заболевание; тест с низкой чувствительностью не идентифицирует многих пациентов с заболеванием, а тест с высокой чувствительностью полезен, чтобы исключить диагноз при отрицательных результатах. Чувствительность является дополнением ложноотрицательных результатов (т. е., ложноотрицательные результаты плюс чувствительность = 100%).

---

<div class = "alert alert-block alert-info">
    
* <b>Специфичность</b>  – представляется собой вероятность отрицательных результатов теста среди пациентов при отсутствии заболевания (истинно-отрицательная частота)

</div>



Таким образом, тест, который отрицателен у 9 из 10 пациентов при отсутствии заболевания имеет специфичность 0,9 (или 90%). Специфичность представляет, насколько хорошо тест правильно идентифицирует пациентов с болезнью, потому что тесты с высокой специфичностью обладают малой долей ложноположительных результатов. Тест с низкой специфичностью диагностирует многих пациентов без заболеваний как имеющих заболевания. Он является дополнением ложноположительных результатов.

## Коэффициент вероятности (КВ)

В отличие от чувствительности и специфичности, которые не относятся к вероятностям конкретного пациента, КВ позволяет интерпретировать результаты теста у конкретного больного при наличии известной (оценочной) предтестовой вероятности болезни.

КВ описывает изменение в претестовой вероятности заболевания, когда результат теста известен и отвечает на вопрос:

> «Насколько послетестовая вероятность теперь изменилась, когда результат теста известен?». 

Многие клинические тесты дихотомические:
* они либо выше точки отсечки (положительные)
* либо ниже пороговой точки (отрицательные) и есть только 2 возможных результата.

Другие тесты дают результаты, которые непрерывны или происходят в диапазоне, в котором выбрано несколько пороговых точек. Фактическая послетестовая вероятность зависит от величины КВ (которая зависит от рабочих характеристик теста) и оценки предтестовой вероятности заболевания. Когда проведенный тест дихотомичен и результат может быть либо положительным, либо отрицательным, чувствительность и специфичность могут быть использованы, чтобы вычислить положительные КВ (КВ+) или отрицательные КВ (КВ-).

<div class = "alert alert-block alert-info">

* <b>КВ+</b> : Коэффициент вероятности положительного результата теста, наблюдаемого у пациентов с болезнью (истинно положительного), к вероятности положительного результата теста у пациентов без заболеваний (ложноположительного)
    
$$\operatorname{КВ+} = \frac{\operatorname{SENS}}{1 - \operatorname{SPEC}}$$    

* <b>КВ-</b> : Коэффициент вероятности отрицательных результатов теста у пациентов с болезнью (ложноотрицательный) к вероятности отрицательных результатов теста у пациентов без заболеваний (истинно отрицательный)
    
$$\operatorname{КВ-} = \frac{1 - \operatorname{SENS}}{\operatorname{SPEC}}$$        

</div>

## Байессовский подход

<div class = "alert alert-block alert-warning">

* <b>Вероятность, что пациент истинно больной : $P(\text{true_1})$</b>

* <b>Результат анализа, что пациент болен : $P(\text{res_1})$</b>

* <b>Результат ML-модели, что пациент болен : $P(\text{ML_1})$</b>
    

</div>


_**Что мы знаем?**_

* По статистике результатов анализа, мы можем получить оценку : $P(\text{res_1})$ и $P(\text{res_0})$
* Точность прибора : $P(\text{true_1} | \text{res_1})$ и $P(\text{true_0} | \text{res_0})$
* Оценка вероятности истинно болен или здоров: 
$$P(\text{true_1}) = P(\text{true_1} | \text{res_1}) P(\text{res_1}) + 
P(\text{true_1} | \text{res_0}) P(\text{res_0})$$

$$P(\text{true_0}) = P(\text{true_0} | \text{res_1}) P(\text{res_1}) + 
P(\text{true_0} | \text{res_0}) P(\text{res_0})$$
