```python
from copy import deepcopy

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_score,\
                            recall_score, roc_auc_score, roc_curve, cohen_kappa_score, fbeta_score
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.cross_decomposition import PLSRegression, PLSCanonical
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression


from sklearn.model_selection import train_test_split, KFold, StratifiedKFold

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

import warnings
warnings.filterwarnings('ignore')
```

## Загрузим данные

```python
samples = pd.read_excel(u'20.01.23 список образцов Моча.xlsx', header=2)
samples.dropna(subset=['Gender'], inplace=True)

samples.fillna(0, inplace=True)
samples.reset_index(drop=True, inplace=True)
```

```python
spectra1 = pd.read_csv('urine_chm001-250.csv')
spectra2 = pd.read_csv('urine_chm251-500.csv')
spectra_init = pd.concat([spectra1, spectra2],  ignore_index=True)

spectra = spectra_init.filter(regex='^Ch')

mmscaler = MinMaxScaler(feature_range=(-1,1))
spectra = pd.DataFrame(mmscaler.fit_transform(spectra.T).T)
del(spectra1, spectra2)

frame_lol = samples[['Dataset', 'TOTAL_a']]
spectra['Dataset'] = spectra_init['Dataset']
spectra = pd.merge(left=spectra, right=frame_lol, how='left', on='Dataset')
```

## Binarized TOTAL_a + PLS ( LDA ) + LogReg

```python
def metrics_ret(y_true, y_pred, y_prob):
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='macro')
    # roc_auc = roc_auc_score(y_true, y_prob)
    cohen_kappa = cohen_kappa_score(y_true, y_pred)
    
    return [acc, f1, cohen_kappa]

def to_cat(y):
    if y==0: return 0
    elif y==1 or y==2: return 1
    elif y>=3 : return 2

spectra.dropna(inplace=True)

X = spectra.iloc[:,:36]
y = pd.Series(list(map(to_cat, spectra['TOTAL_a'])))
LDA = LinearDiscriminantAnalysis(n_components=2)
LDA_spectra = pd.DataFrame(LDA.fit_transform(X,y))
LDA_spectra['target'] = y.values


plt.figure(figsize=(15,6))
plt.title('Distribution')
plt.xlabel('LDA Component')

plt.scatter(LDA_spectra[LDA_spectra.target == 0][0], LDA_spectra[LDA_spectra.target == 0][1],
            label='Normal: {}'.format(y.value_counts().values[0]))
plt.scatter(LDA_spectra[LDA_spectra.target == 1][0], LDA_spectra[LDA_spectra.target == 1][1],
            label='One anomaly: {}'.format(y.value_counts().values[1]))
plt.scatter(LDA_spectra[LDA_spectra.target == 2][0], LDA_spectra[LDA_spectra.target == 2][1],
            label='More anomaly: {}'.format(y.value_counts().values[2]))
plt.legend()
plt.grid()



Metrics = []
FPR, TPR = [], []
skf = StratifiedKFold(n_splits=10)
for train_inds, test_inds in skf.split(X,y):
    Train_X, Test_X = LDA_spectra.iloc[train_inds,:-1], LDA_spectra.iloc[test_inds,:-1]
    Train_y, Test_y = LDA_spectra.iloc[train_inds,-1], LDA_spectra.iloc[test_inds,-1]
    log_reg = LogisticRegression(C=10).fit(Train_X, Train_y)
    Preds = log_reg.predict(Test_X)
    Probs = log_reg.predict_proba(Test_X)
    Metrics.append(metrics_ret(Test_y, Preds, Probs[:,1]))
    print(classification_report(Test_y, Preds))


Metrics = pd.DataFrame(Metrics, columns=['accuracy', 'f-score', 'Cohen`s kappa'])
```
