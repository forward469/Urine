```python
samples = pd.read_excel(u'20.01.23 список образцов Моча.xlsx', header=2)
samples.dropna(subset=['Gender'], inplace=True)

samples.fillna(0, inplace=True)
samples.reset_index(drop=True, inplace=True)
```

```python
spectra1 = pd.read_csv('urine_chm001-250.csv')
spectra2 = pd.read_csv('urine_chm251-500.csv')
spectra_init = pd.concat([spectra1, spectra2],  ignore_index=True)

spectra = spectra_init.filter(regex='^Ch')

mmscaler = MinMaxScaler(feature_range=(0,1))
# spectra = pd.DataFrame(mmscaler.fit_transform(spectra.T).T)
del(spectra1, spectra2)

frame_lol = samples[['Gender','Age','Dataset']+list(samples.filter(regex=r'_a$').columns)]
spectra['Dataset'] = spectra_init['Dataset']
spectra = pd.merge(left=spectra, right=frame_lol, how='left', on='Dataset')
spectra.dropna(subset=samples.filter(regex=r'_a$').columns, inplace=True)
```

```python
spectra = spectra.groupby('Dataset').mean()
```

```python
Spectra_1 = mmscaler.fit_transform(spectra.iloc[:,:36:2].values.T).T
Spectra_2 = mmscaler.fit_transform(spectra.iloc[:,1:36:2].values.T).T
```

```python
def HarmAvg(spec1, spec2, beta=0.5):
    return (1+beta**2)*spec1*spec2/(beta**2*spec1 + spec2)
```

```python
HarmAvg_Spetra = HarmAvg(Spectra_1, Spectra_2, beta=1)
```

```python
plt.plot(HarmAvg_Spetra[0])
plt.plot(HarmAvg_Spetra[1])
plt.plot(HarmAvg_Spetra[2])
```

```python
def get_point_cloud(Spectra, window_width=9, window_step=1):
    PointCloud = []
    window_pos = 0
    while window_pos <= Spectra.shape[0] - window_width:
        PointCloud.append(Spectra[window_pos : window_pos+window_width].tolist())
        window_pos += window_step
    return np.array(PointCloud)
```

```python
WINDOW_WIDTH = 3
WINDOW_STEP = 1

Topology_Spectras = np.array([])

for spec in HarmAvg_Spetra:
    Topology_Spectras = np.append(Topology_Spectras, 
                                  get_point_cloud(spec, 
                                                  window_width=WINDOW_WIDTH,
                                                  window_step=WINDOW_STEP))
```

```python
Topology_Spectras = Topology_Spectras.reshape((-1,(18-WINDOW_WIDTH)//WINDOW_STEP + 1, WINDOW_WIDTH))
```

```python
from sklearn.decomposition import PCA
```

```python
Number_of_Pacients = Topology_Spectras.shape[0]
Number_of_SpectraPoints = Topology_Spectras.shape[1]

PCA_Topology_Spectra = PCA(n_components=3).fit_transform(Topology_Spectras.reshape((-1, 9)))

Topology_Data = pd.DataFrame(PCA_Topology_Spectra, index=spectra.index.values.repeat(10))
Topology_Data = pd.merge(Topology_Data, spectra.filter(regex='_a$'), left_index=True, right_index=True)
```
